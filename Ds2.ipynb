```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# Show first few rows
print("First 5 rows:")
print(iris_df.head())
```

```python
# Data info
print("\nDataset Info:")
print(iris_df.info())

# Summary statistics
print("\nDescriptive Statistics:")
print(iris_df.describe())
```

```python
# Set Seaborn theme
sns.set(style="whitegrid")

# Plot distribution for each numerical column
num_cols = iris_df.columns[:-1]  # exclude species

for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(iris_df, x=col, kde=True, hue='species', element='step')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Scatter plot matrix with species as hue
sns.pairplot(iris_df, hue="species", diag_kind="kde", height=2.5)
plt.suptitle("Pairwise Scatter Plots", y=1.02)
plt.show()
```
